```python
import numpy as np
import uproot
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
ClockDDC10 = 6e8
adccperVolt = 8192
resistance_ohm = 50
sampleWidth_ns = 10
# filter parameters
A = 1.0
B = 0.5
n = 1
m = n
fWidth = 2*n + m
fmask = np.concatenate((B*np.ones(shape=n),-A*np.ones(shape=m),B*np.ones(shape=n)))
```

```python
from numba import jit
#@jit(nopython=True) # Set "nopython" mode for best performance, equivalent to @njit
def Read_DDC10_BinWaveCap(fName, doTime=False):
    waveInfo = {}
    fp = open(fName+'.bin','rb')
    Header = np.fromfile(fp,dtype=np.uint32,count=4)
    
    waveInfo['numEvents'] = int(Header[0])
    waveInfo['numSamples'] = int(Header[1])
    waveInfo['chMap'] = np.array([1 if digit=='1' else 0 for digit in bin(Header[2])[2:]])
    waveInfo['numChan'] = len(waveInfo['chMap'])
    byteOrderPattern = hex(int(np.fromfile(fp,dtype=np.uint32,count=1)))
    
    tmpArr = np.concatenate((np.fromfile(fp,dtype=np.int16),np.zeros(2)))
    fp.close()
    
    waveArr = np.reshape(tmpArr.astype(dtype=np.float64)/adccperVolt,(waveInfo['numEvents'],waveInfo['numChan'],(waveInfo['numSamples']+6)))[...,2:-4]
    
    if doTime:
        with open(fName+'.log','r') as fl:
            waveInfo['liveTimes_s'] = np.loadtxt(fl,delimiter=',',skiprows=5,max_rows=waveInfo['numEvents'],usecols=(2),dtype=np.float64)/(ClockDDC10)
            waveInfo['totliveTime_s'] = np.sum(waveInfo['liveTimes_s'])
    return (waveArr,waveInfo)
```

```python
waveArr,waveInfo = Read_DDC10_BinWaveCap("testdata/test",True)
print(waveArr.shape)
waveInfo
```

```python
peaks = np.argmax(np.absolute(waveArr[:,0,:]),axis=1)
nBins = 350
hpeaks = np.histogram(peaks,bins = nBins,range=(0,350))
plt.plot(np.arange(0,350,350/nBins),hpeaks[0]/waveInfo['totliveTime_s'],marker='+')
#plt.xlim((270,310))
```

```python
#baseline and rms
nBaseSamps = 200
baseWave = waveArr[...,:nBaseSamps]
waveBaseline = np.sum(baseWave,axis = 2)/nBaseSamps
waveBaserms = np.sqrt(np.sum(baseWave*baseWave,axis = 2)/nBaseSamps - waveBaseline*waveBaseline)
#total event charge
tmpWave = waveArr[:,0,:] - waveBaseline
waveQBase = 1e3*integrate.simps(tmpWave[:,:nBaseSamps],axis=1)*sampleWidth_ns/resistance_ohm
waveQ_pC = 1e3*np.sum(tmpWave[:,275:300],axis=1)*sampleWidth_ns/resistance_ohm
#len(waveInfo['liveTimes_s'])

baseMask = np.absolute(waveQBase*np.reciprocal(nBaseSamps*waveBaserms.flatten()))<2

waveBaserms
```

```python
plt.figure()
amin = -2
amax = 10
goodWaveQ = waveQ_pC[baseMask]
Qhist = np.histogram(waveQ_pC,1000,(amin,amax),weights=np.ones(shape=waveQ_pC.shape)/waveInfo['totliveTime_s'])
Qhist2 = np.histogram(goodWaveQ,1000,(amin,amax),weights=np.ones(shape=goodWaveQ.shape)/waveInfo['totliveTime_s'])

plt.plot(Qhist[1][:-1],Qhist[0])
#plt.plot(Qhist2[1][:-1],Qhist2[0])
plt.xlabel('Q [pC]')
plt.ylabel('Rate [Hz]')
plt.xlim((-1,5))
plt.show()
print('Integrated Rate = '+str(integrate.simps(Qhist[0]))+' Hz')
```

```python
#@jit(nopython=True) # Set "nopython" mode for best performance, equivalent to @njit
def numbaConFilter(x,mask):
    #expects x is jagged array
    res = np.zeros(x.shape)
    for i in range(len(x)):
        tmp = np.convolve(x[i],mask,mode="valid")
        res[i] = np.concatenate((np.zeros(len(mask)-1),tmp))
    return res

waveArr[:,0,:]
```
