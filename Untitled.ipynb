```python
import numpy as np
import uproot
import matplotlib.pyplot as plt
from scipy import integrate
import AnaUtils as au
import Triggered as tr
```

```python
wdir = '/data/share/SA5446/Gain_1202V_200617_100_samples_20000_events/0'
waves = au.ReadDDC10_BinWave(wdir)
waves
```

```python
nbase=30
waves[0],base = au.Subtract_Baseline(waves[0],nBase=nbase)
bmask = np.absolute(integrate.simps(waves[0][:,1,:nbase]))<3*integrate.simps(np.ones(nbase))*base[1][:,1]
```

```python
TrigPeaks = au.peakHist(waves,chan=0,ret=True)[1:]
wStart = TrigPeaks[0]+10
wFin = TrigPeaks[0]+30
tmask = TrigPeaks[1]>100
```

```python
evmask = bmask*tmask
```

```python
Qhist = au.winQHist(waves,ch=1,init=wStart.astype(int),end=wFin.astype(int),nBins=50,evMask=evmask[...,np.newaxis])
```

```python
plt.plot(Qhist['qHist'][1],Qhist['qHist'][0])
#plt.yscale('log')
#plt.xlim(-2,10)
plt.xlabel('Q [pC]')
plt.ylabel('Rate [Hz]')
```
