```python
import numpy as np
import uproot
import matplotlib.pyplot as plt
import matplotlib
import AnaUtils as au
import Triggered as tr
import glob,os
```

```python
wDir = '/data/share/KA0219/UV/GdLSDilution/Run1/'
fWaves = [os.path.splitext(iF)[0] for iF in glob.glob(wDir+'/*.bin')]
fWaves.sort()
fWaves,wDir[22:].split('/')
print(fWaves)

waveArrs = [au.ReadDDC10_BinWave(iRun) for iRun in fWaves]
for wave in waveArrs:
    if wave is None:
        waveArrs.remove(None)
waveArrs

sumWaves = [wave[0][:,1,:].sum(0)/float(wave[1]['numEvents']) for wave in waveArrs]
sumWaves = [sWave - sWave[:30].sum()/30.0 for sWave in sumWaves]
#sumWaves

#matplotlib.rc('xtick',labelsize=10)
#matplotlib.rc('ytick',labelsize=10)
for iw in range(len(sumWaves)):
    mark = '+'
    if 'GdLS' in fWaves[iw]:
        mark = 'x'
    plt.plot(sumWaves[iw],label=fWaves[iw][47:52]+'%',marker=mark)
plt.legend(title='Concentration',title_fontsize=12,fontsize=12)
plt.xlabel('Sample (10ns)',fontsize=15)
plt.ylabel('V',fontsize=15)
plt.title('GdLS Dilution (200mL Volume)',fontsize=18)
#plt.xlim(80,500)

#plt.yscale('log')
plt.savefig('UV_AvRun1.png',bbox_inches='tight')
```

```python
intCharge=[]
for i in range(len(sumWaves)):
    charge = 0
    for point in sumWaves[i]:
        charge += point*10
    intCharge.append(charge/50)
    
x=[0.,.005,.01,.05,.1,1,5]
#print(x)
ampOverCharge=[]
for i in range(len(sumWaves)):
    maxC=max(sumWaves[i])/50
    ampOverCharge.append(maxC/intCharge[i])

plt.plot(x, ampOverCharge)
plt.xlabel('GdLS Concentration (%)',fontsize=15)
plt.xscale('log')
plt.ylabel(r'$\mathrm{Signal\,Amplitude/Integrated\,Charge\,(10^9/s)}$',fontsize=15)
plt.title('Signal Amplitude/Integrated Charge (200mL Volume)',fontsize=18)
plt.savefig('UV_ScrRun1.png',bbox_inches='tight')

plt.show()
```
