```python
import numpy as np
import uproot
import matplotlib.pyplot as plt
from scipy import integrate
import AnaUtils as au
import Triggered as tr
import glob,os
```

```python
wDir = '/data/share/KA0219/UV/1780V_100DI_200804_5000_samples_20000_events'
fWaves = [os.path.splitext(iF)[0] for iF in glob.glob(wDir+'/*.bin')]
fWaves.sort()
fWaves
```

```python
waveArrs = [au.ReadDDC10_BinWave(iRun) for iRun in fWaves]
for wave in waveArrs:
    if wave is None:
        waveArrs.remove(None)
```

```python
sumWaves = [wave[0][:,1,:].sum(0)/float(wave[1]['numEvents']) for wave in waveArrs]
sumWaves = [sWave - sWave[:30].sum()/30.0 for sWave in sumWaves]
#sumWaves
```

```python
for iw in range(len(sumWaves)):
    mark = '+'
    if 'GdLS' in fWaves[iw]:
        mark = 'x'
    plt.plot(sumWaves[iw],label=os.path.basename(fWaves[iw]),marker=mark)
plt.legend()
plt.xlabel('Sample (10ns)')
plt.ylabel('V')
#plt.xlim(80,500)

plt.yscale('log')
plt.savefig('UV_Av.png')
```

```python
areas = [au.winQHist(wave,ch=0,init=100,end=5000,sub=True,nBase=50)['qHist'] for wave in waveArrs]
for iw in range(len(areas)):
    mark = '+'
    if 'GdLS' in fWaves[iw]:
        mark = 'x'
    plt.plot(areas[iw][1],areas[iw][0],label=os.path.basename(fWaves[iw]),marker=mark)
plt.legend()
plt.xlabel('Q (pC)')
plt.ylabel('R [Hz]')
#plt.xlim(80,500)

plt.yscale('log')
plt.savefig('UV_Q.png')
```
